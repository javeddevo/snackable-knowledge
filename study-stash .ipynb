- **API**: A way for software components to communicate; can use any protocol (HTTP, gRPC, SOAP, local libraries, OS calls).  
- **REST API**: A lightweight type of API that always uses HTTP and follows REST rules (GET, POST, PUT, DELETE).



- **Local**: Use `uvicorn` to run FastAPI (e.g., `uvicorn main:app --reload`).  
- **AWS Lambda (ZIP deployment)**: Do not import/run `uvicorn`; instead wrap FastAPI with `Mangum` handler.  
- **AWS Lambda (Docker image)**: Uvicorn works fine because the container runs as a server process.
